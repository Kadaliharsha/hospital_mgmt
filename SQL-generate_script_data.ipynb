```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
 
# Initialize Faker and set seeds
fake = Faker()
Faker.seed(42) # Using a different seed for billing data if desired
np.random.seed(42)
random.seed(42)
 
# Number of billing records
n_billing = 300
 
# Read existing patient and appointment data
try:
    patients_df = pd.read_csv("patients_dataset.csv")
    appointments_df = pd.read_csv("appointments_dataset.csv")
 
    # Convert date columns to datetime objects
    patients_df['admission_date'] = pd.to_datetime(patients_df['admission_date'])
    appointments_df['date'] = pd.to_datetime(appointments_df['date']) # Appointment date might have time depending on previous step
 
    patient_ids = patients_df['patient_id'].tolist()
 
    # Create mappings for admission dates and total consulting charges per patient
    patient_admission_dates = patients_df.set_index('patient_id')['admission_date'].to_dict()
 
    # Calculate the total consulting charge for each patient from appointments
    # Group appointments by patient_id and sum the consulting_charge
    patient_total_consulting_charges = appointments_df.groupby('patient_id')['consulting_charge'].sum().to_dict()
 
 
    # Ensure there are enough patient IDs to sample from
    if not patient_ids:
        print("Error: patients_dataset.csv is empty or missing 'patient_id' column.")
        patient_ids = [f"{random.randint(1, 10000)}" for _ in range(n_billing)] # Fallback
        patient_admission_dates = {id: datetime(2023, 1, 1) for id in patient_ids} # Fallback dates
        patient_total_consulting_charges = {} # No charges in fallback
 
 
except FileNotFoundError:
    print("Error: patients_dataset.csv or appointments_dataset.csv not found. Generating random IDs and dates.")
    patient_ids = [f"{random.randint(1, 10000)}" for _ in range(n_billing)] # Fallback
    patient_admission_dates = {id: datetime(2023, 1, 1) for id in patient_ids} # Fallback dates
    patient_total_consulting_charges = {} # No charges in fallback
except KeyError:
    print("Error: Required columns ('patient_id', 'admission_date', 'date', or 'consulting_charge') not found in CSVs. Generating random IDs and dates.")
    patient_ids = [f"{random.randint(1, 10000)}" for _ in range(n_billing)] # Fallback
    patient_admission_dates = {id: datetime(2023, 1, 1) for id in patient_ids} # Fallback dates
    patient_total_consulting_charges = {} # No charges in fallback
 
 
# Generate synthetic data for billing
bill_ids = [f"B{i+1:03d}" for i in range(n_billing)] # Changed padding to 4 digits
chosen_patient_ids_billing = random.choices(patient_ids, k=n_billing) # Randomly select patient IDs for billing
 
billing_dates = []
total_amounts = []
 
for patient_id in chosen_patient_ids_billing:
    admission_date = patient_admission_dates.get(patient_id, datetime(2023, 1, 1)) # Get admission date
 
    # Generate a random date on or after the admission date
    end_date = admission_date + timedelta(days=365) # Billing within a year of admission
 
    delta_days = (end_date - admission_date).days
 
    if delta_days > 0:
        random_days = random.randint(0, delta_days)
        billing_date = admission_date + timedelta(days=random_days)
    else:
        billing_date = admission_date
 
    billing_dates.append(billing_date.strftime("%Y-%m-%d")) # Format date to exclude time
 
    # Get the total consulting charge for this patient
    consulting_charge_for_patient = patient_total_consulting_charges.get(patient_id, 0.0)
 
    # Generate an additional amount for other services (if any)
    # This can be random, but you might want to link it to billed_services later
    other_services_amount = np.round(np.random.uniform(0, 10000), 2) # Random amount for other services
 
    # Calculate the total amount, including consulting charge and other services
    total_amount = consulting_charge_for_patient + other_services_amount
    total_amounts.append(total_amount)
 
 
data = {
    "bill_id": bill_ids,
    "patient_id": chosen_patient_ids_billing,
    "total_amount": total_amounts,
    "billing_date": billing_dates
}
 
df_billing = pd.DataFrame(data)
df_billing.to_csv("billing_dataset.csv", index=False)
 
print("billing_dataset.csv generated successfully!")
```
